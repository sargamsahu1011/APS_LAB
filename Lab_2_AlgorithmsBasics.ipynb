```python
# Exercise 6
# This function sorts the array A using a recursive helper function.
def func8(A):
    func8_helper(A, 0, len(A) - 1)
    return A # Return the sorted array

def func8_helper(A, i, j):
    # Input: An array A and two indices i and j representing the sub-array to sort.
    # This helper function performs a recursive sorting operation on the sub-array A[i...j].
    n = j - i + 1 # Calculate the length of the current sub-array
    if n <= 1: # Base case: if sub-array has 0 or 1 element, it's already sorted
        return
    elif n == 2: # Base case for 2 elements
        if A[i] > A[j]:
            A[i], A[j] = A[j], A[i]
    else: # Recursive step for more than 2 elements
        # The logic here resembles a variation of a ternary sort or a similar recursive pattern.
        # It recursively sorts three overlapping sub-arrays.
        m = n // 3
        func8_helper(A, i, j - m)
        func8_helper(A, i + m, j)
        func8_helper(A, i, j - m)
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(a):
   if not a:
       return None
   min = a[0]
   max= a[0]
   for i in a:
    if i>max:
      max= i
    if i<min:
      min=i
   return min, max
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(a,tar):
  if not a:
    return
  l=0
  h=len(a)-1
  while l<=h:
    mid=(l+h)//2
    if a[mid]==tar:
      return True
    elif a[mid]<tar:
      l=mid+1
    else:
      h=mid-1
      return False
```
