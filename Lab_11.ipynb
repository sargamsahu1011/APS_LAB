# Travelling Salesman Problem using Dynamic Programming.
Problem Description:
Given a set of cities and distances, find the shortest possible route that visits each city once and returns to the starting city.

#Theory
TSP uses dynamic programming with bit masking. Store results of visited subsets to reduce repeated work.

## Algorithm
1. Start from source city
2. Mark visited cities using bit mask
3. Try all unvisited cities
4. Choose minimum cost path
5. Return to source city

```python
def tsp(graph):
    n= len(graph)
    dp= [[-1]*(1<<n)for _ in range(n)]
    parent = [[-1]*(1 << n)for _ in range(n)]

    def visit(c, m):
        if m==(1<<n) - 1:
            return graph[c][0]

        if dp[c][m]!=-1:
            return dp[c][m]

        ans = float('inf')

        for next_c in range(n):
            if m & (1<<next_c) == 0:
                new_cost = graph[c][next_c] + visit(next_c, m| (1<<next_c))
                if new_cost < ans:
                    ans = new_cost
                    parent[c][m] = next_c

        dp[c][m] = ans
        return ans

    return visit(0, 1)
```

```python
graph = [
[0, 10, 15, 20],
[10, 0, 35, 25],
[15, 35, 0, 30],
[20, 25, 30, 0]
]

print("Minimum travelling cost:", tsp(graph))
```

```python

def tsp(graph):
    n = len(graph)

    dp = [[-1] * (1 << n) for _ in range(n)]
    parent = [[-1] * (1 << n) for _ in range(n)]

    def visit(c, m):
        if m == (1 << n) - 1:
            return graph[c][0]

        if dp[c][m]!=-1:
            return dp[c][m]

        ans = float('inf')

        for next_c in range(n):
            if (m & (1 << next_c)) == 0:
                new_cost = graph[c][next_c] + visit(next_c, m | (1 << next_c))

                if new_cost < ans:
                    ans = new_cost
                    parent[c][m] = next_c

        dp[c][m] = ans
        return ans

    min_cost = visit(0, 1)

    # Path Reconstruction
    path = [0]
    m = 1
    c = 0

    while True:
        next_city = parent[c][m]

        if next_city == -1:
            break

        path.append(next_city)
        m |= (1<<next_city)
        c = next_city

    path.append(0)

    return min_cost,path

```

```python
n = int(input("Enter number of cities: "))

graph = []
print("Enter adjacency matrix:")

for i in range(n):
    row = list(map(int, input().split()))
    graph.append(row)
```

```python
cost, path = tsp(graph)

print("Minimum travelling cost:",cost)
print("Path followed:", " -> ".join(map(str, path)))
```
